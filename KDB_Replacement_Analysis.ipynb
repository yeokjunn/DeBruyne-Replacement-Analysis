## 2. Creative Decision Rating (CDR)

One key aspect of KDB we want to replace is his creativity and attacking prowess. To effectively calculate or measure his creativity and attacking output, we use the Creative Decision Rating (or CDR for short).

The Creative Decision Rating (CDR) is a metric used in football analytics to evaluate a player's ability to make effective creative decisions during a match. While there isn't a universally standardized formula for CDR, it typically combines various aspects of a player's performance that contribute to their creative output. How we define the CDR is shown below.

**Weights in CDR:**
1. Progressive Passes per 90  - *20%*

2. Progressive Carries per 90 - *20%*

3. Assists per 90             - *30%*

4. Goals per 90               - *10%*

5. Key passes per 90          - *20%*


### 2.1 Calculating CDR of Manchester City's Midfielders, 2019/20

## 3. Replacement Score (RScore)

To replace KDB, we have to not only take into account the Creative Decision Rating (CDR), but other mitigating factors as well.

The "RScore" represents the potential as a similar replacement for KDB, measured from 0 to 1. The calculation uses a combination of weighted factors to determine the overall RScore for each player.

#### RScore Formula = 
(**CDR** *65%*) + (**League Difficulty** *25%*) +  (**Age** *10%*)

- **Creative Decision Rating (CDR)** - *65%*

Self explanatory, the biggest indicator we have to find our replacement for KDB. As a creative playmaker, KDB's main threat is his creativity and attacking prowess, as mentioned earlier in the CDR section. Hence, the CDR metric carries the majority of the weight at 70% of the overall score.


- **League Difficulty** - *25%*

The difficulty of the league takes into account the competitiveness and reputation of the league in which the player plys his trade in. Players competing in "tougher" leagues will have a higher rating, compared to those plying their trades in weaker leagues. League rating is based on a coefficient from 0 to 1 (1 being the toughest), courtesy of the *UEFA Power Rankings (Wikipedia, 2024)*.


```python
import ssl
from IPython.display import display, Markdown

# code to remove error
ssl._create_default_https_context = ssl._create_unverified_context

UEFA_Ranking_url = "https://en.wikipedia.org/wiki/UEFA_coefficient"
UEFA_Ranking_table = pd.read_html(UEFA_Ranking_url)
UEFA_Ranking_table = UEFA_Ranking_table[1] # select the table I want

# Flatten the MultiIndex columns
UEFA_Ranking_table.columns = [' '.join(col).strip() if isinstance(col, tuple) else col for col in UEFA_Ranking_table.columns]
#print(UEFA_Ranking_table.columns)

# Rename the columns
UEFA_Ranking_table.rename(columns={
    'Ranking 2025': 'Ranking (2024)',
    'Member association (L: League, C: Cup, LC: League Cup) Member association (L: League, C: Cup, LC: League Cup)': 'Nation of League',
    'Coefficient Total': 'Power Coefficient'
}, inplace=True)

# Divide the "Power Coefficient" by 100
UEFA_Ranking_table['Power Coefficient'] = UEFA_Ranking_table['Power Coefficient'] / 100

UEFA_Ranking_table_cleaned = UEFA_Ranking_table[['Ranking (2024)', 
                                                  'Nation of League', 'Power Coefficient']]
UEFA_Ranking_table_cleaned.reset_index(drop=True, inplace=True)

# Remove the (L, C) part from the "Nation of League" column using .loc
UEFA_Ranking_table_cleaned.loc[:, 'Nation of League'] = UEFA_Ranking_table_cleaned['Nation of League'].str.replace(r'\s*\(.*?\)', '', regex=True)


display(Markdown("#### UEFA League Ranking Table (2024)"))
UEFA_Ranking_table_cleaned[:10]
```

- **Age** - *10%*

The age reflects a player's potential for growth, development, and the ability to adapt to the team's requirements (*Worville, 2021*). A younger player will naturally have the ability to become better over experiences & time, whilst an older player might be on the performance decline.

**Score Evaluation:**

| Age Range     | Age Rating |
|:--------------|------:|
| 16 - 21       | 0.80  |
| 22 - 29       | 1.00  |
| 30 - 99       | 0.50  |


 --------------------------------------------------------------------------------------------------------------------------


**Explanation:**

Younger players (16 - 21) have the potential to drop off or improve, which is volatile. This explains the high, but not full rating of 0.80 given, accounting improvement or deprovement over time.

Experienced players (22 - 29) have more or less solidified their ability to produce at that sustained level, which explains the full 1.00 rating given.

Veteran players (30 - ) are given a lower rating (0.50), as it accounts for the possibility that older players may have valuable experience and maturity, but their potential for improvement may be less significant.

```python
# Start code on compiling RScore for top5_midfielders_df_CDR.csv

# reading in csv file
top5_midfielders_df_CDR = pd.read_csv("top5_midfielders_df_CDR.csv") 
top5_AllPlayers_Data = pd.read_csv("top5_AllPlayers_Data.csv")
```

## 5. Euclidean Distance for Multi-Dimensional Similarity

#### Introduction to Euclidean Distance

Euclidean distance measures the "straight-line" distance between two points in a multi-dimensional space. 

Each dimension corresponds to a variable (e.g., "Key Passes per 90", "Goals per 90"). This distance metric is useful for finding the similarity or dissimilarity between data points (e.g., players in this project) across multiple attributes.

The formula for Euclidean distance in *n*-dimensional space is:


\begin{equation} d(p, q) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2} \end{equation}

#### Why Use Euclidean Distance in This Project?

Unlike correlation-based methods that measure linear relationships, Euclidean Distance captures the absolute differences between players by treating each performance metric as a dimension in a multi-dimensional space. 




By normalizing all metrics to a common scale, the Euclidean Distance metric allows for a more holistic comparison, emphasizing overall similarity rather than individual trends. 





This approach is particularly useful for identifying players with performance profiles most closely resembling a target player, such as Kevin De Bruyne, by calculating the shortest distance in the feature space. **Basically, the smaller the distance, the more the player resembles De Bruyne.**
Through this method, we gain deeper insights into player profiles and uncover hidden similarities in their playing styles and contributions.



### 5.1 Using Min-Max Scaler (Normalisation Technique 1)

```python
from scipy.spatial.distance import euclidean

# Load the datasets
data = pd.read_csv("top5_midfielders_df_CDR&RScore.csv")
city_1920_stats = pd.read_csv("city_1920_midfielders_CDR.csv")

# Specify the columns to consider
columns_to_consider = [
    "Minutes Played", "Key Passes per 90",
    "Progressive Carries per 90", "Progressive Passes per 90",
    "Goals per 90", "Assists per 90"
]

# Add Kevin De Bruyne's stats to `data` temporarily for scaling
kevin_stats = city_1920_stats[city_1920_stats["Player"] == "Kevin De Bruyne"]
kevin_stats = kevin_stats[["Player"] + columns_to_consider]  # Include only necessary columns

data_with_kevin = pd.concat([data, kevin_stats], ignore_index=True)
```

```python
# Normalize the combined dataframe
scaler = MinMaxScaler()
data_with_kevin[columns_to_consider] = scaler.fit_transform(data_with_kevin[columns_to_consider])

# Extract normalized Kevin De Bruyne stats
kevin_normalized = data_with_kevin[data_with_kevin["Player"] == "Kevin De Bruyne"][columns_to_consider].iloc[0]

# Remove Kevin De Bruyne from the combined dataset (if present in `data` already)
data_normalized = data_with_kevin[data_with_kevin["Player"] != "Kevin De Bruyne"]

# Compute Euclidean distance for all players
data_normalized["Euclidean Distance"] = data_normalized[columns_to_consider].apply(
    lambda row: euclidean(row, kevin_normalized), axis=1
)

# Sort players by similarity (ascending Euclidean distance)
data_sorted = data_normalized.sort_values(by="Euclidean Distance")

# Display the top results
(data_sorted[["Player", "Euclidean Distance"]].head(20))
```

## 6. K-Means Clustering

#### Introduction to K-Means Clustering

K-Means clustering is an unsupervised machine learning algorithm used to group data into clusters based on their similarity. 

Each cluster is represented by a centroid, and the algorithm iteratively adjusts the centroids to minimize the variance within clusters. In the context of this project, it can be used to group players into clusters based on their multi-dimensional statistics, such as "Key Passes per 90", "Goals per 90", and other attributes.

#### Why Use K-Means Clustering in This Project?

1. **Identifying Similarities Across Multiple Metrics**



K-Means allows us to evaluate players holistically by taking multiple metrics into account (e.g., "Key Passes per 90," "Progressive Passes per 90").

Players with similar playing styles and performance metrics will naturally group together, making it easier to find replacements or compare them against each other.

2. **Flexibility in Analysis**


K-Means is flexible because you can adjust the number of clusters to fine-tune the analysis.
Small 𝑘 allows for broader groupings.
Large  𝑘 allows for more granular distinctions between players.
We are able to experiment with different values of 𝑘 to achieve the level of detail you need for player comparison.

3. **Insights Beyond Simple Metrics**


Methods like Euclidean distance or Pearson correlation focus on pairwise comparisons, which might miss patterns across larger groups.
K-Means examines the overall dataset structure and creates groups based on the intrinsic properties of the data.

#### 6.1.1 Cluster K = 5

```python
from sklearn.cluster import KMeans

# Apply K-Means clustering
k = 5  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
data_with_kevin3['Cluster'] = kmeans.fit_predict(data_with_kevin3[columns_to_consider])
```

```python
# Highlight the target player (e.g., Kevin De Bruyne)
target_player = "Kevin De Bruyne"
target_player_data = data_with_kevin3[data_with_kevin3["Player"] == target_player]
plt.scatter(
    target_player_data["Key Passes per 90"], 
    target_player_data["Goals per 90"], 
    color="red", label="Kevin De Bruyne", s=200, edgecolor="black"
)

plt.title("K-Means Clustering of Players (K = 5)")
plt.xlabel("Key Passes per 90")
plt.ylabel("Goals per 90")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display the cluster containing Kevin De Bruyne
target_cluster = target_player_data['Cluster'].values[0]
similar_players = data_with_kevin3[data_with_kevin3['Cluster'] == target_cluster]
(similar_players[["Player", "Cluster"]])
```
